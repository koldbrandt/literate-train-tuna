```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_res_trans = pd.read_csv('../trainstats/Resnet50-transfer.csv', index_col=0)
df_res = pd.read_csv('../trainstats/Resnet50-no-transfer.csv', index_col=0)
df_cnn_batch = pd.read_csv('../trainstats/CNN_batch_norm.csv', index_col=0)
df_basic_adam = pd.read_csv('../trainstats/Basic_adam.csv', index_col=0)
df_basic_sgd = pd.read_csv('../trainstats/Basic_sgd.csv', index_col=0)
```

```python
df_res_trans.head()
```

```python
df_res.head()
```

```python
# plot test acc
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))
ax1.plot(df_res_trans['train_acc'], label ='Resnet50-Transfer' )
ax1.plot(df_res['train_acc'], label = 'Resnet50' )
ax1.plot(df_cnn_batch['train_acc'] , label = 'CNN Batchnorm')
ax1.plot(df_basic_adam['train_acc'] , label = 'Basic Adam')
ax1.plot(df_basic_sgd['train_acc'] , label = 'Basic SGD')
ax1.set_ylabel('Train Accuracy')
ax1.set_xlabel('Epoch')
ax2.plot(df_res_trans['test_acc'], label ='Resnet50-Transfer' )
ax2.plot(df_res['test_acc'], label = 'Resnet50' )
ax2.plot(df_cnn_batch['test_acc'] , label = 'CNN Batchnorm')
ax2.plot(df_basic_adam['test_acc'] , label = 'Basic Adam')
ax2.plot(df_basic_sgd['test_acc'] , label = 'Basic SGD')
ax2.set_ylabel('Test Accuracy')
ax2.set_xlabel('Epoch')
plt.legend()
plt.show()
```

```python
# subplot with test loss and train loss
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))
ax1.plot(df_res_trans['train_loss'], label ='Resnet50-Transfer' )
ax1.plot(df_res['train_loss'], label = 'Resnet50' )
ax1.plot(df_cnn_batch['train_loss'] , label = 'CNN Batchnorm')
ax1.plot(df_basic_adam['train_loss'] , label = 'Basic Adam')
ax1.plot(df_basic_sgd['train_loss'] , label = 'Basic SGD')
ax1.set_ylabel('Train Loss')
ax1.set_xlabel('Epoch')
ax2.plot(df_res_trans['test_loss'], label ='Resnet50-Transfer' )
ax2.plot(df_res['test_loss'], label = 'Resnet50' )
ax2.plot(df_cnn_batch['test_loss'] , label = 'CNN Batchnorm')
ax2.plot(df_basic_adam['test_loss'] , label = 'Basic Adam')
ax2.plot(df_basic_sgd['test_loss'] , label = 'Basic SGD')
ax2.set_ylabel('Test Loss')
ax2.set_xlabel('Epoch')
plt.legend()
plt.show()
```
